### Model Integration with groq

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["OPENROUTER_API_KEY"]=os.getenv("OPEN_ROUTER_API")
```

```python
from langchain.chat_models import init_chat_model

model=init_chat_model('groq:llama-3.3-70b-versatile')
model
```

```python
resonse = model.invoke("what is todays date")
resonse.content
```

```python
from langchain_groq import ChatGroq

modell = ChatGroq(model='llama-3.3-70b-versatile')
resonsee = modell.invoke("what is today date")
resonsee.content
```

### Streaming

```python
for chunk in modell.stream("Write me a two hundred words on artifical intellegence"):
    print(chunk.text, end="", flush=True)
```

```python
for chunk in modell.stream("Why do parrots have feather"):
    print(chunk.text, end="", flush=True)
```

### Batching

```python
responseee = modell.batch(
    [
    "what is todays date",
    "How do airplanes fly",
    "What is quantum physics"]
)

for response in responseee:
    print(response)
```

```python
ressponse= modell.batch(
    [
    "what is todays date",
    "How do airplanes fly",
    "What is quantum physics"],
    config={
        "max_currency": 5
    }
)

for response in ressponse:
    print(response)
```
