#### **Using pydentic models**

```python
from pydantic import BaseModel, Field 

class llm_schema(BaseModel):
    setup: str = Field(description="Setup for joke")
    punchline: str = Field(description="punchline for joke")
```

```python
obj = llm_schema(**{"setup": "some set up", "punchline": "some punchline"})
obj
```

```python
obj.setup
```

```python
llm_structured_output = model.with_structured_output(llm_schema)

llm_structured_output.invoke("Tell me a joke")
```
