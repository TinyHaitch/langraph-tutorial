### **MESSAGES** 

```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")
my_msg = [
    SystemMessage(content="youre a gen-z assistant that always answer in a fun way"),
    HumanMessage(content="Bro tell me a funfact")
]

model.invoke(my_msg).content
```

### **PROMPT**

```python
from langchain_core.prompts import PromptTemplate

user_question = input("What do you wanna ask? ")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic": user_question})

model.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a polite assistant"),
    ("user", "Write a funfact about {topic}")
])

ready_prompt = prompt_template.invoke({"topic": "AI"})
ready_prompt.messages
```
