### ***Chain with Custom runnable***

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
#### TASK-1
prompt_template = ChatPromptTemplate([
    ("system", "You are a study assistant"),
    ("human", "{input}")
])
```

```python
#### TASK-2
from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
```

```python
#### TASK-3

str_passout = StrOutputParser()
```

```python
#### TASK-4
from langchain_core.runnables import RunnableLambda

def dictionary_maker(text:str) -> dict:
    return{"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)
```

```python
#### TASK-5

prompt_post = ChatPromptTemplate(
   messages = [
        ("system", "You are a social media post generator"),
        ("human", "create a post for the following text for linkedIn {text}")
    ]
)
```

```python
#### TASK-6
model = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
```

```python
#### TASK-7

str_passout = StrOutputParser()
```

```python
#### CHAIN

chain = prompt_template | model | str_passout | dictionary_maker_runnable | prompt_post | model | str_passout

chain.invoke("What is the capical of france")
```
